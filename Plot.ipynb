```python
import numpy as np
import seaborn
%matplotlib inline
#Все функции matplotlib API, в частности plot и close, находятся в модуле matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
def format_table(source):
    table = np.loadtxt(source)
    
    formatted_table = np.zeros((201, 201))
    for i in range(200, -1, -1):
        for j in range(0, 201, 1):
            formatted_table[i][j] = table[202 * i + j][2]
            
    return formatted_table

def pic(formatted_table, output):
    fig = plt.figure(figsize=(16,12))
    seaborn.set(font_scale=2)
    h = seaborn.heatmap(formatted_table, cmap="RdBu_r", square=True, linewidths=0.0)
    h.set_xticks([1, 51, 101, 151, 201])
    h.set_xticklabels(labels=[-10, -5, 0, 5, 10], fontdict={'fontsize': 30})
    h.set_yticks([1, 51, 101, 151, 201])
    h.set_yticklabels(labels=[-10, -5, 0, 5, 10], fontdict={'fontsize': 30})

    h.tick_params(length=10, width=3)

    cax = plt.gcf().axes[-1]
    cax.tick_params(labelsize=20, length=10, width=3)
    cax.xaxis.label.set_size(20)

    fig.savefig(output, dpi=1000)
```

```python
source_list = ["../build-Custom_Omnivehicle_Control_Predict-Desktop_Qt_5_14_0_GCC_64bit-Debug/energy_0.dat",
              "../build-Custom_Omnivehicle_Control_Predict-Desktop_Qt_5_14_0_GCC_64bit-Debug/energy_turn.dat"]

output_list = ["./energy_0.png", "./energy_turn.png"]

for i in range(len(source_list)):
    pic(format_table(source_list[i]), output_list[i])
```
